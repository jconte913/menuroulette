```python
import import_ipynb
import Menus
%run ./Menus.ipynb

import random

week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Define generic options
generic_meals = ["Leftovers", "Date Night"]

# Meal type slots
meal_slots = {
    "generic": 2,       # 2 days for generic meals
    "pasta": 1,         # 1 pasta day
    "fish": 1,          # 1 fish day
    "intensive": 1,     # 1 intensive meal
    "easy": 2           # 2 easy/quick meals
}

# Randomly assign days for each slot
available_days = week.copy()
meal_plan = {}

# Generic days
generic_days = random.sample(available_days, meal_slots["generic"])
for day in generic_days:
    meal_plan[day] = random.choice(generic_meals)
    available_days.remove(day)

# Pasta day
pasta_day = random.choice(available_days)
meal_plan[pasta_day] = random.choice(list(pastas.keys()))
available_days.remove(pasta_day)

# Fish day
fish_day = random.choice(available_days)
meal_plan[fish_day] = random.choice(list(fish.keys()))
available_days.remove(fish_day)

# Intensive meal day
intensive_day = random.choice(available_days)
meal_plan[intensive_day] = random.choice(list(intensive_meals.keys()))
available_days.remove(intensive_day)

# Easy meals for remaining days
for day in available_days:
    meal_plan[day] = random.choice(list(easy_meals.keys()))

# Lunch plan
chosen_lunch = random.choice(lunch_options)
lunch_plan = {day: chosen_lunch for day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]}

# Build shopping list
shopping_list = []
for meal in meal_plan.values():
    ingredients = (
        easy_meals.get(meal)
        or intensive_meals.get(meal)
        or pastas.get(meal)
        or fish.get(meal)
    )
    if ingredients:
        shopping_list.extend(ingredients)

# Print meal plans
print("\nYour Weekly Meal Plan:\n")
for day in week:
    print(f"{day}: {meal_plan[day]}")

print("\nWeekday Lunch Plan:")
for day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]:
    print(f"{day}: {lunch_plan[day]}")

print("\nShopping List:")
for item in shopping_list:
    print(f"- {item}")
```
